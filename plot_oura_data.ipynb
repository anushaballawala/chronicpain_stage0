```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/userdata/aallawala/pain_data/oura/oura_sleep_2025-06-02T17-39-23.csv')

print(df.columns)
```

```python


# Convert 'day' to datetime format
df['day'] = pd.to_datetime(df['day'])
df = df.sort_values('day')
```

```python
df['day']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert to hours
df['total_sleep_hours'] = df['total_sleep_duration']/3600

# Optional: sort by day
df = df.sort_values('day')

plt.figure(figsize=(14, 6))
sns.boxplot(x='day', y='total_sleep_hours', data=df, color='lightblue')
sns.stripplot(x='day', y='total_sleep_hours', data=df, color='black', alpha=0.5, jitter=True)

plt.xticks(rotation=45)
plt.title('Total Sleep Hours  Box Plots Over Time')
plt.ylabel('Sleep Duration (hours)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
# Convert to hours
df['deep_sleep_duration'] = df['deep_sleep_duration']/3600

# Optional: sort by day
df = df.sort_values('day')

plt.figure(figsize=(14, 6))
sns.boxplot(x='day', y='deep_sleep_duration', data=df, color='lightblue')
sns.stripplot(x='day', y='deep_sleep_duration', data=df, color='black', alpha=0.5, jitter=True)

plt.xticks(rotation=45)
plt.title('Deep sleep duration')
plt.ylabel('Sleep Duration (hours)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
